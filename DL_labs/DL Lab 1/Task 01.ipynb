```python
import numpy as np
import matplotlib.pyplot as plt

# Generate distributions
exp_data = np.random.exponential(scale=1.0, size=100000)
uni_data = np.random.uniform(low=0.0, high=5.0, size=100000)
norm_data = np.random.normal(loc=2.0, scale=1.0, size=100000)

# Plot histograms with labels
plt.hist(exp_data, bins=100, density=True, histtype='step', label='Exponential', color='red')
plt.hist(uni_data, bins=100, density=True, histtype='step', label='Uniform', color='green')
plt.hist(norm_data, bins=100, density=True, histtype='step', label='Normal', color='blue')

# Add legend after plotting
plt.legend(loc='upper right')

# Additional formatting
plt.axis([0, 8, 0, 1.2])
plt.title("Comparison of Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)

# Show the plot
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Create X, Y meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

# 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.viridis)
ax.set_title("Z = X² + Y²")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Load the dataset to df dataframe
df = pd.read_csv('/content/drive/MyDrive/DL_Lab/Pokemon.csv', encoding='ISO-8859-1')

# Rename if needed to match lowercase format
df.rename(columns={
    'HP': 'hp', 'Attack': 'attack', 'Defense': 'defense',
    'Sp. Atk': 'sp_attack', 'Sp. Def': 'sp_defense', 'Speed': 'speed'
}, inplace=True)

# Select required features
features = ['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']
df = df[features].dropna()

# Pearson correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(method='pearson'), annot=True, cmap='coolwarm')
plt.title("Pearson Correlation Heatmap")
plt.show()

# Spearman correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(method='spearman'), annot=True, cmap='YlGnBu')
plt.title("Spearman Correlation Heatmap")
plt.show()
```
